```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\devad\Downloads\laptop_prices.csv")
```

```python
avg_price_by_inches = data.groupby('Inches')['Price_euros'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_price_by_inches.index, avg_price_by_inches.values, marker='o', color='blue')
plt.title('Average Price by Screen Size (Inches)')
plt.xlabel('Screen Size (Inches)')
plt.ylabel('Average Price ($)')
plt.grid()
plt.show()
```

```python
avg_price_by_ram = data.groupby('Ram')['Price_euros'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_price_by_ram.index, avg_price_by_ram.values, marker='x', color='green')
plt.title('Average Price by RAM Size (GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Average Price (€)')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Weight'], data['Price_euros'], marker='^', linestyle='-', color='red', alpha=0.6)
plt.title('Price vs Weight of Products')
plt.xlabel('Weight (kg)')
plt.ylabel('Price (€)')
plt.grid()
plt.show()
```

```python
avg_price_by_type = data.groupby('TypeName')['Price_euros'].mean()

plt.figure(figsize=(12, 6))
plt.plot(avg_price_by_type.index, avg_price_by_type.values, marker='s', color='purple')
plt.title('Average Price by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Price (€)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
